```python
import matplotlib.pyplot as plt


lista_tempo = []
lista_posicao = []
tempo = 0.0
dt = 1.0
posicao = 0.0
while tempo < 10:
  posicao_inicial = 0
  velocidade = 3
  posicao_tempo = posicao_inicial + velocidade * tempo
  tempo = tempo + dt
  lista_tempo.append(tempo)
  lista_posicao.append(posicao_tempo)

plt.xlabel('$t$ ($s$)')
plt.ylabel('$posição$ (m)')
plt.plot(lista_tempo, lista_posicao)
plt.title('MRU')
```

```python
import numpy as np
import matplotlib.pyplot as plt

#variaveis
x=0.0
y=0.0
x0=0
y0=0
t=0
grav=10
v0=10
ang=60
a=0.5
##converção para radianos
radAng=np.radians(ang)


# Tempo de simulação
tempo = 2 * v0 * np.sin(radAng) / grav
t = 100  # Número de passos de tempo
deltaT = np.linspace(0, tempo, t)



vx=v0*np.cos(radAng)
x=x0+vx*deltaT

vy=v0*np.sin(radAng)
y=vy*deltaT-((grav*deltaT**2)/2)

# Plotar a trajetória
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Trajetória')
plt.xlabel('Distância (m)')
plt.ylabel('Altura (m)')
plt.title('Lançamento oblíquo sem arrasto')
plt.show()
```
